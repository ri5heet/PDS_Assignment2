**e.)Perform select, filter, rename, mutate, arrange and summarize with group by operations (or**
**their equivalent operations in python) on this dataset**.

```python
# Load the dataset
df = data_clean

# Group by 'Location' and 'Fuel_Type'
grouped_df = df.groupby(['Name', 'Year', 'Fuel_Type'])

# Selecting specific columns after grouping
selected_grouped_df = grouped_df[['Name', 'Year', 'Fuel_Type']]

# Filtering groups based on a condition (e.g., total Price > 5000000)
filtered_grouped_df = grouped_df.filter(lambda x: x['Price'].sum() > 15)
fgdf=filtered_grouped_df.to_csv('/content/filter_grouping.csv',index=False)


# Renaming columns
renamed_grouped_df = grouped_df.agg({'Price': 'mean', 'Seats': 'max'}).rename(columns={'Price': 'Average_Price'})
rgdf=renamed_grouped_df.to_csv('/content/rename_grouping.csv',index=False)


# Mutating: Add a new column 'Price_Per_Kilometer' (Price divided by Kilometers_Driven) for each group
mutated_grouped_df = grouped_df.apply(lambda x: x.assign(Price_Per_Kilometer=x['Price'] / x['Kilometers_Driven']))
mgdf=mutated_grouped_df.to_csv('/content/mutate_grouping.csv',index=False)

# Arranging rows within each group by 'Year' in descending order
arranged_grouped_df = grouped_df.apply(lambda x: x.sort_values(by='Year', ascending=False))
agdf=arranged_grouped_df.to_csv('/content/arranged_grouping.csv',index=False)


# Summarizing: Calculate the mean 'Price' and maximum 'Seats' for each group
summary_grouped_df = grouped_df.agg({'Price': 'mean', 'Seats': 'max'})
sgdf=summary_grouped_df.to_csv('/content/summary_grouping.csv',index=False)


print("Selected Columns for Grouped Data:")
print(selected_grouped_df.head())

print("\nFiltered Grouped Data:")
print(filtered_grouped_df.head())

print("\nRenamed Columns for Grouped Data:")
print(renamed_grouped_df.head())

print("\nMutated Grouped Data:")
print(mutated_grouped_df.head())

print("\nArranged Grouped Data by Year:")
print(arranged_grouped_df.head())

print("\nSummary Statistics for Grouped Data:")
print(summary_grouped_df)
```
